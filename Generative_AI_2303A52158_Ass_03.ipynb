**1. (1 ponto) Write Python code without using any libraries to find the value of x at which the
function f(x) shown in equation (1) has minimum value. Consider Gradient Descent Algorithm.
f (x) = 5x4 + 3x2 + 10**

```python
def f(x):
    return 5 * x**4 + 3 * x**2 + 10

def df(x):
    return 20 * x**3 + 6 * x

x = 1.0
learning_rate = 0.01
max_iters = 1000
tolerance = 1e-6

for _ in range(max_iters):
    grad = df(x)
    new_x = x - learning_rate * grad

    if abs(new_x - x) < tolerance:
        break

    x = new_x

print("Minimum value of x:", x)
print("Minimum function value f(x):", f(x))
```

**2. (1 ponto) Write Python code without using any libraries to find the value of x and y at which the
function g(x,y) shown in equation (2) has minimum value. Consider Gradient Descent Algorithm.
f (x) = 3x2 + 5e−y + 10**

```python
def g(x, y):
    return 3 * x**2 + 5 * (2.71828**-y) + 10

def dg_dx(x):
    return 6 * x

def dg_dy(y):
    return -5 * (2.71828**-y)

x, y = 1.0, 1.0
learning_rate = 0.01
max_iters = 1000
tolerance = 1e-6

for _ in range(max_iters):
    grad_x = dg_dx(x)
    grad_y = dg_dy(y)
    new_x = x - learning_rate * grad_x
    new_y = y - learning_rate * grad_y

    if abs(new_x - x) < tolerance and abs(new_y - y) < tolerance:
        break

    x, y = new_x, new_y

print("Minimum values:", "x:", x, "y:", y)
print("Minimum function value g(x, y):", g(x, y))
```

**4. (1 ponto) Write Python code without using any libraries to find the value of optimal values of
model parameters M and C such that the model’s Square Error Value shown in equation 4 will
be minimum. It means model gives output close to expected output as shown in Figure 1**

```python
# Initial values
M, C = 1.0, 1.0
data_points = [(1, 2), (2, 2.8), (3, 3.6), (4, 4.5)]
learning_rate = 0.01
max_iters = 1000
tolerance = 1e-6

for _ in range(max_iters):
    grad_M = sum(-2 * x * (y - (M * x + C)) for x, y in data_points) / len(data_points)
    grad_C = sum(-2 * (y - (M * x + C)) for x, y in data_points) / len(data_points)

    new_M = M - learning_rate * grad_M
    new_C = C - learning_rate * grad_C

    if abs(new_M - M) < tolerance and abs(new_C - C) < tolerance:
        break

    M, C = new_M, new_C

print("Optimal values: M =", M, ", C =", C)
```
